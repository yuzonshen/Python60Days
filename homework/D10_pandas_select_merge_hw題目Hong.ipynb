題目:<br>
讀取STOCK_DAY_0050_202009.csv串聯STOCK_DAY_0050_202010.csv，並且找出open小於close的資料

```python
import pandas as pd
```

```python
#讀取STOCK_DAY_0050_202009.csv、STOCK_DAY_0050_202010.csv
stock2009 = pd.read_csv('STOCK_DAY_0050_202009.csv')
print(stock2009)
stock2010 = pd.read_csv('STOCK_DAY_0050_202010.csv')
print(stock2010)
```

```python
#串聯
concatData = pd.concat([stock2009,stock2010],axis=0)
print(concatData)

#合併
mergeData = pd.merge(stock2009,stock2010,on='date',how='outer')
print(mergeData)
```

```python
#找出open小於close的資料
```

```python
result = mergeData.loc[(mergeData.open_x < mergeData.close_x) | (mergeData.open_y < mergeData.close_y),['date','open_x','close_x','open_y','close_y']]
print(result)
print('len=',len(result))
```

```python
resultb = concatData.loc[concatData.open<concatData.close,['date','open','close']]
print(resultb)
print('len=',len(resultb))
```
