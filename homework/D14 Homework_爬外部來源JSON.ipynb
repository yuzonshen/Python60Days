### 2. 請將 Dcard API 取得所有的看板資訊轉換成 DataFrame，並且依照熱門程度排序後存成一個 csv 的檔案。


```python
import requests
r = requests.get('https://www.dcard.tw/_api/forums')
response = r.text

import json
data = json.loads(response)
#print(data)
#print(type(data))

#轉換成 DataFrame
df3 = pd.DataFrame(data)
df3
```

```python
#依照熱門程度排序後存成一個 csv 的檔案
df_sorted = df3.sort_values(by=['subscriptionCount'], ascending=False)

#df_sorted.to_csv('DcardAPI_sorted.csv')
df_sorted.to_excel('DcardAPI_sorted.xlsx', sheet_name='sorted')
```
