# 範例重點

如何使用亂數, 資料集來操作


```python
# 載入需要的...
import matplotlib.pyplot as plt
import numpy as np

# 準備數據 ... 假設我要畫一個sin波 從0~180度
# 設定要畫的的x,y數據list....
x = np.arange(0,180)
y = np.sin(x * np.pi / 180.0)

# 開始畫圖
plt.plot(x,y)
# 在這個指令之前，都還在做畫圖的動作 
# 這個指令算是 "秀圖"
plt.show() 

#畫出完整的 sin 圖形

x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)

plt.plot(x, y_sin)

plt.show()


#畫出完整的 cos 圖形
# 畫出 cos 圖檔，並儲存
x = np.arange(0, 3 * np.pi, 0.1)
y = np.cos(x)

plt.plot(x,y)
#plt.show()
plt.savefig("filename.png",dpi=300,format="png")
```

# 散點圖: Scatter Plots
    
    顏色由（X，Y）的角度給出。

    注意標記的大小，顏色和透明度。

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

T = np.arctan2(X, Y)
plt.scatter(X,Y, s=100, c=T,alpha=0.6)
plt.title("Scatter Plot")
plt.show()
```

```python
# 1D Array constructed Randomly
random_array = np.random.rand(5)
print("1D Array filled with random values : \n", random_array)

plt.plot(random_array)
plt.show()
```

```python
# 1D Array as per Gaussian Distribution 高斯分布
mean = 0
std = 0.1
array = np.random.normal(mean, std, 1000)
#print("1D Array filled with random values "
#      "as per gaussian distribution : \n", array)

#plt.plot(array)
#plt.show()

count, bins, ignored = plt.hist(array, 30, normed=True)
#print (count)

plt.plot(bins,1/(std * np.sqrt(2 * np.pi)) *
          np.exp( - (bins - mean)**2 / (2 * std**2) ),
         linewidth=2, color='r')
plt.show()
```
