### Q1:計算標準常態分配，小於1的機率有多大?
![](Q1.png)

```python
# library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
import seaborn as sns
```

```python
## Q1
#  提示: 用 cdf 來算

#1.設定使用參數
mu=0
sigma=1
x=np.linspace(-10, 10, 500)
# 2.產生 pdf圖
y = stats.norm.pdf(x,mu, sigma)
plt.plot(x, y, "b-", linewidth=2)
#plt.grid(True)
plt.title('pdf of Norma(0,1)')
plt.show()
# 3.計算常態分配的累積機率 (cumulative density function)，pdf 的積分
cumsum_probs =  stats.norm.cdf(x,mu, sigma)
plt.ylabel('P(X<=x)')
plt.xlabel('x')
plt.title(' cdf of Nomal(0,1)')
plt.plot(x, cumsum_probs,"b-", linewidth=2)
plt.show()

#### 計算標準常態分配，小於 1 的機率有多大？
x=1
cumsum_probs =  stats.norm.cdf(x,mu, sigma)
print('小於 1 的機率=',cumsum_probs)
```

### Q2:計算標準常態分配，大於1，小於 -1 的機率有多大?
![](Q2.png)


```python
## Q2
x=1
cumsum_probs =  stats.norm.cdf(x,mu, sigma)
print('小於 1 的機率=',cumsum_probs)
x=-1
cumsum_probs_1 =  stats.norm.cdf(x,mu, sigma)
print('小於 -1 的機率=',cumsum_probs_1)
print('在一個標準差間的機率=',cumsum_probs-cumsum_probs_1)
print('大於1，小於 -1 的機率=',1-(cumsum_probs-cumsum_probs_1))
```

```python
#標準常態的特性：
#68.3% 的數值，落在平均數 ±1 個標準差間；
#95.4% 的數值，落在平均數 ±2 個標準差間；
#99.7% 的數值，落在平均數 ±3 個標準差間。
```

### Q3:X~N(2,4),x 服從常態分配，平均數為2,變異數為 4，計算 X小於 6 的機率有多大?
