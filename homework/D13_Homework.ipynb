````python
# 記得先 Import 正確的套件

import pandas as pd
import numpy as np
````

### 1. 請接下列資料依照指定規則做合併：

```
- 依照 fruit 欄位做合併
- 依照 index 索引做合併
```


````python
df1 = pd.DataFrame({
    'fruit': ['apple', 'banana', 'orange'] * 3,
    'weight': ['high', 'medium', 'low'] * 3,
    'price': np.random.randint(0, 15, 9)
})

df2 = pd.DataFrame({
    'fruit': ['apple', 'orange', 'pine'] * 2,
    'weight': ['high', 'low'] * 3,
    'price': np.random.randint(0, 15, 6)
})
print(df1)
print(df2)
````

````python
# - 依照 fruit 欄位做合併

pd.merge(df1,df2, on='fruit')
````

````python
# - 依照 index 索引做合併

df1.join(df2,lsuffix='_left',rsuffix='_right')
````

### 2. [簡答題] 承上題，請問為什麼依照 merge 合併後有些資料不見了？


````python
# merge 預設: 交集  ; how可指定聯集 (outer join or left outer, right outer)

pd.merge(df1,df2, on='fruit', how='outer')

#pd.merge(df1,df2, on='fruit', how='left')

#pd.merge(df1,df2, on='fruit', how='right')
````

### 3. [簡答題] 承上題，請問為什麼依照 index 合併會發生錯誤？請用程式解決。

````python
名稱重複:　因為join不像merge會自動對於重複欄位產生字尾，所以需要參數lsuffix、rsuffix加上指定字尾
````

````python
df1.join(df2,lsuffix='_df1',rsuffix='_df2')
````
