```python
# import library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
import seaborn as sns
from IPython.display import display
```

## 讀入資料

```python
df_train = pd.read_csv("Titanic_train.csv")
df_train.info()
print(df_train['Age'].isnull().sum())
display(df_train.head(5))
```

## 觀察 Age

```python
#step1: 觀察 Age 和 Pclass 與 Sex 是否有關連性? =>Sex男性平均年齡高於女性, Pclass1的平均年齡高於>Pclass2>Pclass3,呈遞減關係 
#plt.scatter(x="Age", y="Pclass",data=df_train)
g = sns.catplot(x="Sex", y="Age", hue="Pclass", data=df_train, kind="box")
```

```python
g = sns.catplot(x="Pclass", y="Age", data=df_train, kind="box")
```

```python
g = sns.catplot(x="Sex", y="Age", data=df_train, kind="box")
```

```python
#step2: 如果有關連性，運用 KNN ，取出 Age 、 Pclass、 Sex的資料，以 Sex 與 Pclass 補 Age 遺失值。
# 離散轉連續型資料
#透過 preprocessing 轉換
from sklearn import preprocessing
#下面兩行程式，讓 Male=1, FeMale-0.
le = preprocessing.LabelEncoder()
df_train['Sex']=le.fit_transform(df_train['Sex'])
display(df_train.head(5))
```

```python
#計算資料點的倆倆距離
from sklearn.metrics.pairwise import nan_euclidean_distances
display(pd.DataFrame(nan_euclidean_distances(df_train.loc[:,['Sex','Pclass','Age']])))
```

```python
#透過 KNN 進行補值
# KNN
# 設定 k 值
display(df_train[df_train['Age'].isnull().values==True])

value_neighbors=3
from sklearn.impute import KNNImputer
#初始化: Initialize KNNImputer
imputer = KNNImputer(n_neighbors=value_neighbors,add_indicator= True)
df_filled = pd.DataFrame(imputer.fit_transform(df_train.loc[:,['Sex','Pclass','Age']]))
display(df_filled[df_filled.isnull().values==True])
df_filled.rename(columns={0:'Sex',1:'Pclass',2:'Age',3:'imputed_Age'},inplace=True)
display(df_filled)
display(df_filled[df_filled['imputed_Age']==1])
df_filled[df_filled['imputed_Age']==1].to_excel('D38_fill_train.xlsx')
```
